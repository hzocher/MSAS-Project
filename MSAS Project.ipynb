```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Trevor Lawrence NFL Stats.csv")
```

```python
df.plot('Rk',"Rate")
plt.xlabel('Game')
plt.ylabel('Passer Rating')
plt.title('Trevor Lawrence NFL Passer Rating Over Time')
```

```python
df['5_game_group']=np.arange(len(df))//5
```

```python
df.groupby('5_game_group')['Rate'].mean().plot()
plt.xlabel('Game (5 game intervals)')
plt.ylabel('Passer Rating')
plt.title('Trevor Lawrence NFL Passer Rating Over Time')
```

```python
df['3_game_group']=np.arange(len(df))//3
```

```python
df.groupby('3_game_group')['Rate'].mean().plot()
plt.xlabel('Game (3 game intervals)')
plt.ylabel('Passer Rating')
plt.title('Trevor Lawrence NFL Passer Rating Over Time')
```

```python
df.groupby('3_game_group')['QBR'].mean().plot()
plt.xlabel('Game (3 game intervals)')
plt.ylabel('QBR')
plt.title('Trevor Lawrence NFL QBR Over Time')
```

```python
plt.scatter(df.Rk, df.Rate)
a, b = np.polyfit(df.Rk, df.Rate, 1)
plt.plot(df.Rk, a*df.Rk+b)
plt.xlabel('Game')
plt.ylabel('Passer Rating')
plt.title('Trevor Lawrence NFL Passer Rating Over Time')
plt.show()
print(df.Rk.corr(df.Rate))
print(a)
print(b)
```

```python
plt.scatter(df.Rk, df.QBR)
c, d = np.polyfit(df.Rk, df.QBR, 1)
plt.plot(df.Rk, c*df.Rk+d)
plt.xlabel('Game')
plt.ylabel('QBR')
plt.title('Trevor Lawrence NFL QBR Over Time')

plt.show()
print(df.Rk.corr(df.QBR))
print(c)
print(d)
```

```python
plt.scatter(df.Rk, df.EPA)
a, b = np.polyfit(df.Rk, df.EPA, 1)
plt.plot(df.Rk, a*df.Rk+b)
plt.xlabel('Game')
plt.ylabel('EPA')
plt.title('Trevor Lawrence NFL EPA Over Time')
plt.show()
print(df.Rk.corr(df.EPA))
print(a)
print(b)
```
